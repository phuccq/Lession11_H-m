```python
import turtle
import math
t = turtle.Turtle()
t.speed(0)

def pen_up(x,y):
    """Hàm này dùng để nhấc bút lên và di chuyển đến tọa độ cần vẽ"""
    t.penup()
    t.goto(x,y)
    t.pendown()
    
def draw_rectangular(x,y,c):
    """Hàm này dùng để vẽ hình chữ nhật từ 2 cạnh x, y được nhập vào từ người dùng và tô màu c cho hình"""
    t.setheading(0)    
    t.fillcolor(c)
    t.begin_fill()
    
    for i in range(2):
        t.forward(x)
        t.right(90)
        t.forward(y)
        t.right(90)
        
    t.end_fill()
    t.setheading(0)    

def draw_triangle(x,c):
    """Hàm này dùng để vẽ hình tam giác vuông cân từ cạnh đáy x được nhập vào từ người dùng"""    
    t.setheading(0)
    t.fillcolor(c)
    t.begin_fill()
    
    t.forward(x)
    t.left(135)
    t.forward(math.sqrt(x*x/2))
    t.left(90)
    t.forward(math.sqrt(x*x/2))
    
    t.end_fill()
    t.setheading(0)

def draw_trapezium(x,y,z):
    """Hàm này dùng để vẽ hình thang cân từ cạnh đáy x, cạnh bên y và góc đáy z"""
    t.forward(x)
    t.left(180-z)
    t.forward(y)
    t.left(z)
    t.forward((y*math.sin(math.radians(90-z)))/math.sin(math.radians(90)))
    t.left(z)
    t.forward(y)

def draw_tree(a,b):
    """Hàm này dùng để vẽ cây có độ rộng thân cây a; độ dài cạnh đáy tán là lớn nhất b nhập vào từ người dùng"""
    # Vẽ thân cây có chiều rộng bằng a và chiều cao bằng 3a
    draw_rectangular(a,a*3,"brown")
    # Di chuyển đến tọa độ vẽ tán lá lớn nhất
    t.backward((b-a)/2)
    # Vẽ tán lá lớn nhất
    draw_triangle(b,"green")
    # Di chuyển đến tọa độ vẽ tán lá lớn thứ 2, 
    t.penup()
    t.forward((b-b*0.75)/2)
    t.left(90)
    t.forward(b/2)
    t.pendown()
#     # tọa độ x1 = tọa độ x - 1/2 chiều rộng thân cây + 1/2 cạnh đáy tán lá thứ 2; 
#     # tọa độ y1 = tọa độ y + chiều cao thân cây + chiều cao tán lớn nhất (=1/2 cạnh đáy tán thứ 1)
#     x2 = x-a/2+(b*0.75/2)
#     y2 = y+b/2
#     pen_up(x2,y2)
    # Vẽ tán lá lớn thứ 2
    draw_triangle(b*0.75,"green")
    # Di chuyển đến tọa độ vẽ tán lá nhỏ nhất
    t.penup()
    t.forward((b-b*0.75)/2)
    t.left(90)
    t.forward(b*0.75/2)
    t.pendown()
#     # tọa độ x3 = tọa độ x1 + 1/2 phần chênh giữa cạnh đáy tán thứ 1 và cạnh đáy tán thứ 2; 
#     # tọa độ y3 = tọa độ y1 + 1/2 cạnh đáy tán nhỏ nhất
#     x3 = x2 - b*0.25/2
#     y3 = y2 + b*0.5
#     pen_up(x3,y3)
    # Vẽ tán lá lớn nhỏ nhất
    draw_triangle(b*0.5,"green")
    
# Vẽ cây nhỏ
pen_up(-200,-100)
draw_tree(20,160)
# Vẽ nhà
    # Vẽ tường nhà
pen_up(-80,120)
draw_rectangular(300,150,"yellow")
    # Vẽ cửa ra vào
pen_up(40,70)
draw_rectangular(60,100,"blue")
    # Vẽ ống khói
pen_up(50,200)
draw_rectangular(20,80,"grey")
    # Vẽ mái nhà
pen_up(-80,120)
draw_triangle(200,"red")
# Vẽ cây to
pen_up(350,-120)
draw_tree(40,260)

turtle.done()
```
